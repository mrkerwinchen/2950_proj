```python
rawdata.groupby("decade").count()
```

```python
rawdata = rawdata[["decade", "year", "artist", "song", "lyrics"]]
```

```python
rawdata
```

```python
for i in rawdata.index:
    rawdata.loc[i, "lyrics"] = re.sub(r'(\[.*?\]|\,|\.|\!|\?|\"|\(|\)|\{|\})', '', rawdata["lyrics"][i]).lower()
    rawdata.loc[i, "lyrics"] = re.sub(r'\n', ' ', rawdata["lyrics"][i]).strip()
    rawdata.loc[i, "lyrics"] = re.sub(r'\binstrumental\b', '', rawdata["lyrics"][i]).strip()
```

```python
rawdata = rawdata.reset_index()
```

```python
rawdata = rawdata[["decade", "year", "artist", "song", "lyrics"]]
```
