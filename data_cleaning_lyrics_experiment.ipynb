```python
import pandas as pd
import numpy as np
import re
import lyricsgenius as lg
```

```python
raw6012 = pd.read_csv("6012.csv")
raw1319 = pd.read_csv("1319.csv")
total = raw6012.append(raw1319).reset_index()
total = total[["decade", "year", "artist", "song"]]
total.to_csv("totalsongs_6019.csv")
```

```python
total.loc[41, "song"] = "Mister Custer"
```

```python
for i in total.index:
    total.loc[i, "artist"] = re.sub(r' feat.*', '', total["artist"][i])
    total.loc[i, "song"] = re.sub(r' \/.*', '', total["song"][i])
```

```python
total["lyrics"] = pd.DataFrame(["None"] * 6000)
```

```python
total["genius_year"] = pd.DataFrame([0] * 6000)
```

```python
total
```

```python
progress = 100
genius = lg.Genius("JAlCQvWQxOy0Ertp8NhDj4wHzxBwc12vQiToA2HkRFqDHCLOBYTj0DjgrZCldg0f", sleep_time=0.01, verbose = False)
for i in total.index:
    try:
        locate_song = genius.search_song(total["song"][i], total["artist"][i])
        total.loc[i, "lyrics"] = locate_song.lyrics
    except:
        total.loc[i, "lyrics"] = "NONE" #if error thrown (ie there is no lyrics), put "NONE" into the lyrics list
    try:
        total.loc[i, "genius_year"] = locate_song.year
    except:
        total.loc[i, "genius_year"] = 9999
    if i % progress == 0:
        print("Done with", i, "songs") #allows me to track the progress of the scraping
        progress += 100
```

```python
total.to_csv("total_lyrics.csv")
```
