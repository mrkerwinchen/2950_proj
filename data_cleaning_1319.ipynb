```python
raw = pd.read_csv("2013raw.csv", names = ["mix"])
```

```python
artist = []
song = []
for i in raw["mix"]:
    try:
        artist.append(i[i.index(".")+1:i.index("–")].strip())
        song.append(i[i.index("– ")+1:].strip())
    except:
        print(i)
```

```python
raw1719 = pd.read_csv("1719.csv", names = ["year", "mash"])
raw1719 = raw1719.dropna()
raw1719 = raw1719.reset_index()
```

```python
year = []
artist = []
song = []
i = 0
while i <= len(raw1719) - 3:
    #print(raw1719.loc[i]["year"])
    year.append(raw1719.loc[i]["year"])
    artist.append(raw1719.loc[i+2]["mash"])
    song.append(raw1719.loc[i+1]["mash"])
    i += 3
```

```python
temp1719 = pd.DataFrame({"decade":[2010]*300, "year":year, "artist":artist, "song":song})
```

```python
clean = clean.append(temp1719)
len(clean)
```

```python
import lyricsgenius as lg
```

```python
lyrics = []
genius = lg.Genius("JAlCQvWQxOy0Ertp8NhDj4wHzxBwc12vQiToA2HkRFqDHCLOBYTj0DjgrZCldg0f")
progress = 100
for i in range(len(clean)):
    locate_song = genius.search_song(clean.song[i], clean.artist[i])
    try:
        lyrics += [locate_song.lyrics]
    except AttributeError:
        lyrics += ["NONE"] #if error thrown (ie there is no lyrics), put "NONE" into the lyrics list
    if i % progress == 0:
        print("Done with", i, "songs") #allowed me to track the progress of the scraping
        progress += 100
print("Done with all songs")
```

```python
clean['lyrics'] = lyrics
```
